Facebook's forecasting method for time series, https://facebookincubator.github.io/prophet/docs/quick_start.html#python-api

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet

%matplotlib inline

import matplotlib
```

```python
df = pd.read_csv('examples/example_wp_peyton_manning.csv')
df['y'] = np.log(df['y'])
print(df.count())
df.head()
```

```python
m = Prophet()
m.fit(df);
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
