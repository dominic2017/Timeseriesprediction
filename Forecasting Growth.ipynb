```python
import pandas as pd
import numpy as np
from fbprophet import Prophet

%matplotlib inline

import matplotlib
```

```python
df = pd.read_csv('examples/example_wp_R.csv')
import numpy as np
df['y'] = np.log(df['y'])
```

```python
# We must specify the carrying capacity in a column cap. Here we will assume a particular value, but this 
# would usually be set using data or expertise about the market size.
```

```python
df['cap'] = 8.5
```

```python
m = Prophet(growth='logistic')
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=1826)
future['cap'] = 8.5
fcst = m.predict(future)
m.plot(fcst);
```
